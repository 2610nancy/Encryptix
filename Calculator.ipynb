```python
import customtkinter
import tkinter as tk

# Initialize appearance mode
customtkinter.set_appearance_mode('dark')

# Create the main window
janela = customtkinter.CTk()
janela.title("Calculator")
janela.geometry('400x500')  # Adjusted size for better layout

def calcular():
    try:
        calculo = output.get('0.0', 'end').strip()
        resultado = eval(calculo)
        output.delete('0.0', 'end')
        output.insert('0.0', resultado)
    except Exception as e:
        output.delete('0.0', 'end')
        output.insert('0.0', 'Error')

def insert_text(text):
    output.insert('end', text)

def clear_output():
    output.delete('0.0', 'end')

# Create output display
output = customtkinter.CTkTextbox(janela, width=380, height=80, corner_radius=10, border_width=5, border_color='#042940', font=('Arial', 30))
output.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

# Button creation helper function
def create_button(text, command, row, column, colspan=1):
    button = customtkinter.CTkButton(janela, text=text, command=command, corner_radius=10, width=80, height=60, font=('Arial', 20))
    button.grid(row=row, column=column, padx=5, pady=5, columnspan=colspan, sticky="nsew")

# Define button layout
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0, 2), ('.', 4, 2), ('+', 4, 3),
    ('C', 5, 0, 1), ('=', 5, 1, 3)
]

# Create buttons
for button in buttons:
    if len(button) == 3:
        text, row, column = button
        create_button(text, lambda t=text: insert_text(t), row, column)
    elif len(button) == 4:
        text, row, column, colspan = button
        if text == 'C':
            create_button(text, clear_output, row, column, colspan)
        elif text == '=':
            create_button(text, calcular, row, column, colspan)
        else:
            create_button(text, lambda t=text: insert_text(t), row, column, colspan)

# Configure row and column weights
for i in range(6):
    janela.grid_rowconfigure(i, weight=1)
for j in range(4):
    janela.grid_columnconfigure(j, weight=1)

# Run the application
janela.mainloop()
```
